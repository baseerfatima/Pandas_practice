# MELT

```python
import pandas as pd
```

```python
var=pd.DataFrame({"First":['a','b','a','b','a', 'a' ], "s1":[1,2,3,4,5,6], 's2':[7,8,9,0,8,5]})
var
```

```python
pd.melt(var)
```

```python
pd.melt(var,id_vars=["First"])
```

```python
pd.melt(var,id_vars=["First"],var_name="This", value_name="valval")
```

# PIVOT

```python
var=pd.DataFrame({"First":['a','b','a','b','a', 'a' ], "s1":[1,2,3,4,5,6], 's2':[7,8,9,0,8,5], "sec":[10,20,30,40,50,60]})
var
```

```python
var.pivot(index="First", columns="sec")
```

```python
var.pivot(index="First", columns="sec", values="s1")
```

```python
var.pivot_table(index="First",columns="sec", aggfunc="mean",margins="True")
```

```python
var.pivot_table(index="First",columns="sec", aggfunc="sum",margins="True")
```
